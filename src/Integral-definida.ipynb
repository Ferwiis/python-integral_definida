## **Descripción del trabajo**

En el presente trabajo se calculará la integral definida (a, b) de diferentes funciones, tales como: 
1. f(x)= Sen(x)+1
2. f(x)= Cos(x)+2
3. f(x)= e^x-1
4. f(x)= Ln(x)

A partir de las expresiones anteriores se calculará el área bajo la curva, con distintos valores tanto en sus intervalos como el número de sus divisiones (n). 
Para las integrales se van a presentar dos método: 
1. Método rectangulo: 
2. Método trapecio

Para plantear la función de la integral se evalua un condicional que pide que b sea mayor que a, con intervalos que pueden estar dentro de rango de 3 a 2000. Se realiza el código determinando que dentro de el se pueda mirar que el argumento del primer parámetro (sen, cos, ln, e) cumpla las siguientes condiciones: 

1. Que sen y Cos estén dentro de un rango 0 a 2*pi 
2. Que la exponencial este dentro de un rango de 0 a 8 
3. Por último que el logaritmo natural este dentro de un rango de 1 al 20

No obstante el código contará con que dependiendo del parámetro, lo mandará a una pequeña función (función_a_integrar),que se encargará de crear una nueva lista de valores dependiendo si es (sen, cos, e, Ln) para que pueda irse directamente a las definiciones de las funciones matemáticas, básicamente agrega los desplazamientos que tenga la función (+1, +2, -1) para adiconarlos,  guardarlos y devolverlos a la lista. 

En el código se plantea que si es "rect" hará el calculo por el método de los rectángulos y si es "trap" será por trapecios.

1. Rectágulos: para los rectángulos recorre un ciclo que va hasta n porque es la cantidad de intervalos en la que se dividide, se guarda en la variable área que inicia en 0 el valor en cada posición en i de y1, que es donde se calcula los valores para ese intervalo, multiplicandose por el diferencial dx que viene siendo los pasos. Tomando la fórmula del área del rectángulo área por altura, redondeando el resultado a las 6 cifras. 

2. Trapecio: Para este se tomo la siguiente fórmula: 

&gt; &gt; &gt; $\int_{a}^{b} \  f(x) \, dx= \ h/2 * ( {\ f(a)} + {2f(x1)} + {f(b)} )$

Ahora para el trapecio este también recorre un ciclo pero desde 1 hasta n, desde 1 porque aquí se va a sumar el valor de la primera posición de la lista y llega hasta n porque evalua al final b, es decir, se esta sumando el valor de la función evaluada en a y luego evaluada en b. 
Se almacena los valores que estan en la mitad en el programa, es decir, i=2 hasta n-1, los suma dentro de un ciclo, luego se multiplica por 2. Para completar la formula inicia sumando lo de la mitad porque pueden ser muchos términos y después se guarda en la variable lo de la función evaluada en a f(a) y lo de la función evaluda en b f(b) y procede a multiplicarse por h/2 (altura sobre dos) actuando este como un diferecial para esta ocasión y se redondea a las 6 cifras.

En el caso de las series de Taylor determinará cada función evaluandolo en n valores. Es decir, suma todos los valores hasta n y da el mismo valor de la función, en este caso, manejamos la serie de Taylor para Sen, Cos, e y Ln.

### Trabajo 1. Cálculo de la integral por los métodos

```python
def integral(funcion,a,b,metodo,n):
  '''integral es una función que calcula la integral definida de la función que se indique en el parámetro \'funcion\' por alguno
     de los 2 métodos de integración numérica que maneja, los cuales son: rectángulos y trapecios. En la función se recibirán una
     serie de parámetros que provienen de cada línea del archivo que se está recorriendo en la función leer_archivo y que son 
     traídas a esta función. Si se cumplen con las restricciones que se evalúan en los condicionales y en la función
     determinar_intervalos, entonces se procede a integrar la función; en caso contrario se manda una alerta por incumplimiento con los
     parámetros de entrada. Dependiendo el valor de la variable \'metodo\', se decidirá si se integrará la funcion por rectángulos con el valor \'rect\'
     o trapecios con el valor \'trap'. El valor del área se retornará.
     Entradas
       funcion: Es una variable tipo string que contiene el nombre de una de las 4 funciones que recibe el programa: sen, cos, exp o ln.
       a: Es una variable tipo string que trae un valor numérico que corresponde al limite inferior de integración. Esta variable
       luego será pasada a tipo flotante con el método float() para poder utilizarla en la integración
       b: Es una variable tipo string que trae un valor numérico que corresponde al límite superior de integración. Esta variable
       luego será pasada a tipo flotante con el método float() para poder utilizarla en la integración 
       metodo: Es una variable tipo string que contiene la abreviación de uno de los 2 posibles métodos de integración: \'rect\' (rectángulos) o \'trap\' (trapecios)
       n: Es una variable de tipo entero que guarda el número de rectángulos o trapecios en los que se dividirá el área de la función a integrar
     Salidas
       area: Es una variable numérica de tipo float que contiene el área debajo de la curva (integral definida) según el método de cálculo utilizado (rectángulos o trapecios)
     Error
       * Si a es diferente de un valor númerico (caracteres especiales o letras), la función mostrará un error
       * Si b es diferente de un valor númerico (caracteres especiales o letras), la función mostrará un error
       * Si n es diferente de un valor númerico (caracteres especiales o letras), la función mostrará un error
       * Si el valor de funcion es diferente de alguna de las cadenas (sen, cos, exp, ln), la función mostrará un error
       * Si el valor de metodo es diferente de alguna de las cadenas (rect, trap), la función mostrará un error
  '''
  a = float(a)
  b = float(b)
  n = int(n)
  if b > a and 3<=n<=2000 and determinar_intervalos(funcion,a,b):
    dx = (b-a)/(n)
    valor = a
    x1 = [(a+(dx*i)) for i in range(n+1)]
    y1 = funcion_a_integrar(funcion, x1)
    area = 0
    if metodo == "rect":
      for i in range(n):
        area += ((y1[i])*(dx))
    elif metodo == "trap":
      for i in range(1,n):
        area += y1[i]
      area*=2
      area+=y1[0]+y1[n]
      area*=dx/2
    area = round(area,6)
    return area
  else:
    print("\n-------------------------------------------------------------------------------------------------------------------\nALERTA: La linea en lectura \'"+funcion+","+str(a)+","+str(b)+","+metodo+","+str(n)
    +"\' no será procesada por que incumple uno o varios de sus parámetros\n-------------------------------------------------------------------------------------------------------------------\n")
```

### Trabajo 2. Funciones para el cálculo de las funciones matemáticas.


#### Funcion $n!$

```python
def factorial(n):
  '''factorial es una función que tiene como objetivo calcular el factorial de un número entero n. Al final retorna el valor
     Entradas
       n: Es una variable de tipo entera que representa el número al cual se le busca encontrar el fatorial
     Salidas
       fact: Es una variable de tipo entera que guarda el valor del factorial del número n.
     Error
       * Si n es una variable numérica de tipo flotante, la función mostrará un error
       * Si n es menor a 0 (negativa), la función mostrará un error
  '''
  fact = 1
  for i in range(2,n+1):
    fact*=i
  return fact
```

#### Funcion $sen(x)$

```python
def sen(x):
  '''sen es una función que tiene como objetivo el evaluar un valor x en la función matemática sen(x) y retornar ese valor.La forma en cómo se calcula el seno es mediante su Serie de Taylor
     Entradas
       x: Es una variable de tipo float que representa un valor númerico a evaluar en la función.
     Salidas
       f_x: Es una variable de tipo float que almacena el valor evaluado de x en la función matemática del seno
     Error
       Si x es de un tipo diferente al entero o flotante, se disparará un error
  '''
  n = 16
  f_x = 0
  for i in range(n):
    f_x+=(((-1)**(i))/(factorial(2*i+1)))*(x**(2*i+1))
  return round(f_x,10)
```

#### Funcion $cos(x)$

```python
def cos(x):
  '''cos es una función que tiene como objetivo el evaluar un valor x en la función matemática cos(x) y retornar ese valor. La forma en cómo se calcula el coseno es mediante su Serie de Taylor
     Entradas
       x: Es una variable de tipo float que representa un valor númerico a evaluar en la función.
     Salidas
       f_x: Es una variable de tipo float que almacena el valor evaluado de x en la función matemática del coseno
     Error
       Si x es de un tipo diferente al entero o flotante, se mostrará un error
  '''
  n = 16
  f_x = 0
  for i in range(n):
    f_x+=(((-1)**(i))/(factorial(2*i)))*(x**(2*i))
  return round(f_x,10)
```

#### Funcion $e^x$

```python
def exp(x):
  '''exp es una función que tiene como objetivo el evaluar un valor x en la función matemática exp(x) y retornar ese valor. La forma en cómo se calcula la función exponencial es mediante su Serie de Taylor
     Entradas
       x: Es una variable de tipo float que representa un valor númerico a evaluar en la función.
     Salidas
       f_x: Es una variable de tipo float que almacena el valor evaluado de x en la función matemática exponencial
     Error
       Si x es de un tipo diferente al entero o flotante, se mostrará un error
  '''
  n = 38
  f_x = 0
  for i in range(n):
    f_x+=((x**i)/(factorial(i)))
  return round(f_x,10)
```

#### Funcion $ln(x)$

```python
def ln(x):
  '''ln es una función que tiene como objetivo el evaluar un valor x en la función matemática ln(x) y retornar ese valor. La forma en cómo se calcula el logaritmo natural es mediante su Serie de Taylor
     Entradas
       x: Es una variable de tipo float que representa un valor númerico a evaluar en la función.
     Salidas
       f_x: Es una variable de tipo float que almacena el valor evaluado de x en la función matemática del logaritmo natural
     Error
       Si x es de un tipo diferente al entero o flotante, o es 0 o menor a 0 se mostrará un error
  '''
  n = 49
  f_x = 0
  for i in range(n):
    f_x+=(1/(2*i+1))*(((x-1)/(x+1))**(2*i+1))
  f_x*=2
  return round(f_x,10)
```

### Trabajo 3. Función para probar

```python
def leer_archivo(arch):
  '''leer_archivo es una función que se encarga de leer el archivo de entrada. Esto lo hace extrayendo cada una
     de sus lineas con la función open(). En una lista quedan almacenadas todas las líneas del archivo y posteriormente
     son recorridas en un ciclo for con el cual se extraen cada uno de los parámetros de la línea que son: funcion,a,b,método,n.
     Entradas
       arch: Es el archivo que se ha insertado al Notebook. En específico este archivo lo busca el programa como \'pruebas.txt\'
     Salidas
       Valor del resultado de cada parámetro introducido en el archivo de entrada
     Error
       Si el archivo \'pruebas.txt\' no está cargado en el Notebook se disparará un error
  '''
  instrucciones = arch.readlines()
  arch.close()
  n_ins = int(instrucciones[0])
  instrucciones.pop(0)
  arch_resultados = open("resultados.txt","w")
  for i in range(n_ins):
    items = instrucciones[i].split(',')
    area = integral(*items)
    if n_ins > 1:
      arch_resultados.write(str(area)+"\n")
    else:
      arch_resultados.write(str(area))
    graficar(items[0],float(items[1]),float(items[2]),items[3],area)
  arch_resultados.close()
```

### Trabajo 4. Función que produce la gráfica

```python
def graficar(funcion, a, b, metodo, area):
  '''graficar es una función que tiene la finalidad de realizar la gráfica de la función a integrar,
     incluyendo sus puntos, el área debajo de la curva, el título de la gráfica, los textos (función y area), el limite inferior y superior
     de integración y el dibujo de los rectángulos o trapecios.
     Entradas
       a: Es una variable tipo float que trae un valor numérico que corresponde al limite inferior de integración.
       b: Es una variable tipo float que trae un valor numérico que corresponde al límite superior de integración.
       funcion: Es una variable tipo string que contiene el nombre de una de las 4 funciones que recibe el programa: sen, cos, exp o ln.
       metodo: Es una variable tipo string que contiene la abreviación de uno de los 2 posibles métodos de integración: \'rect\' (rectángulos) o \'trap\' (trapecios)
     Salidas
       N/A
     Error
       N/A
  '''
  x_suavidad = []
  val = a
  while val<=b:
    x_suavidad.append(val)
    val+=(3**-9)
  y_suavidad = funcion_a_integrar(funcion, x_suavidad)
  n_graf = 8
  dx_graf = (b-a)/(n_graf)
  x_graf = [(a+(dx_graf*i)) for i in range(n_graf+1)]
  y_graf = funcion_a_integrar(funcion, x_graf)
  res = "A = "+str(area)
  tit_met = ""
  fig, ax = plt.subplots()
  plt.plot(x_suavidad,y_suavidad,'tab:blue')
  if metodo == "rect":
    for i in range(len(x_graf)-1):
      x_rect = [x_graf[i], x_graf[i], x_graf[i+1], x_graf[i+1], x_graf[i]]
      y_rect = [0, y_graf[i], y_graf[i], 0, 0]
      plt.plot(x_rect,y_rect,'k')
      plt.stackplot(x_rect,y_rect,colors='#ffb4a4')
    tit_met+="rectángulos"
  elif metodo == "trap":
    tit_met+="trapecios"
    for i in range(len(x_graf)-1):
      x_trap = [x_graf[i], x_graf[i], x_graf[i+1], x_graf[i+1], x_graf[i]]
      y_trap = [0, y_graf[i], y_graf[i+1], 0, 0]
      plt.plot(x_trap,y_trap,'k')
      plt.stackplot(x_trap,y_trap,colors='#ffb4a4')
  plt.stem(x_graf,y_graf,'k',basefmt="k")
  plt.ylim(ymin=0)
  plt.title("Aproximación de la integral definida con "+tit_met,fontweight="bold")
  xticks_pos = ax.get_xticks()
  yticks_pos = ax.get_yticks()
  plt.text(xticks_pos[len(xticks_pos)-1],yticks_pos[((len(yticks_pos)-1)//2)-1],res,fontsize=20)
  plt.text(xticks_pos[len(xticks_pos)-1],yticks_pos[(len(yticks_pos)-1)//2],traer_funcion(funcion),fontsize=20)
  dy_ticks = yticks_pos[1]-yticks_pos[0]
  plt.text(a,-1.2*dy_ticks,"a",fontsize=14,fontweight="bold")
  plt.text(b,-1.2*dy_ticks,"b",fontsize=14,fontweight="bold")
  plt.show()
```

### Funciones complementarias indispensables en la ejecución de los trabajos

```python
def traer_funcion(funcion):
  '''traer_funcion es una función cuyo propósito es determinar que función se está leyendo (sen, cos, exp o ln) y con base
     en esto se devuelve una cadena que representa la función escrita formalmente como en Matemáticas bajo el formato \'f(x) = y\' donde \'y\' es
     el valor de la cadena \'funcion\' y la variable que se evalúa (por defecto x) y demás información de la función como el
     desplazamiento en el eje y (f(x) = y + c) u otro tipo de operaciones. En el caso de este programa están establecidas
     funciones (sen(x)+1, cos(x)+2, exp(x)-1 y ln(x)). Según sea el caso la función retornará alguno de estos valores de tipo string.
     Entradas
       funcion: Es una variable tipo string que contiene el nombre de una de las 4 funciones que recibe el programa: sen, cos, exp o ln.
     Salidas
       f_x: Es una variable de tipo string que contiene la modificación del valor de la variable \'funcion\' quedando en formato \'f(x) = y\'
     Error
       N/A
  '''
  f_x = ""
  if funcion == "sen":
    f_x = "f(x) = "+funcion+"(x) + 1"
  elif funcion == "cos":
    f_x = "f(x) = "+funcion+"(x) + 2"
  elif funcion == "exp":
    f_x = "f(x) = "+funcion+"(x) - 1"
  elif funcion == "ln":
    f_x = "f(x) = "+funcion+"(x)"
  return f_x

def determinar_intervalos(funcion,a,b):
  '''determinar_intervalos es una función cuyo propósito es determinar qué función se está leyendo en la línea del archivo
     para con ella definir si los límites de integración para la función matemática determinada están dentro del intervalo
     correspondiente.
     Entradas
       funcion: Es una variable tipo string que contiene el nombre de una de las 4 funciones que recibe el programa: sen, cos, exp o ln.
       a: Es una variable de tipo float que representa el límite inferior de integración
       b: Es una variable de tipo float que representa el límite superior de integración
     Salidas
       verdad: Es una variable de tipo boolean que define si se cumplen las condiciones sobre los límites de integración a y b para
       ser integrada la función en lectura en el archivo.
     Error
       * Si a es diferente de un valor númerico, la función mostrará un error
       * Si b es diferente de un valor númerico, la función mostrará un error
  '''
  verdad = False
  if funcion == "sen":
    pi = 3.1415926535
    verdad = a>=0 and b<=(2*pi)
  elif funcion == "cos":
    pi = 3.1415926535
    verdad = a>=0 and b<=(2*pi)
  elif funcion == "exp":
    verdad = a>=0 and b<=8
  elif funcion == "ln":
    verdad = a>=1 and b<=20
  return verdad


def funcion_a_integrar(funcion, val_x):
  '''funcion_a_integrar es una función que se ocupa de tomar el string del primer parámetro de cada línea del archivo llamado \'funcion\', y mediante condicionales identificar
     la función correspondiente a evaluar para una lista de valores de x. El cálculo de los valores de las funciones se hace mediante listas comprimidas,
     definiendo la funcion que se va a evaluar y asi obtener los valores de y en la lista val_y. Por ejemplo \'val_y = [sen(x)+1 for x in val_x]\' se va hasta la función sen(x) del programa,
     y lleva cada uno de los valores x de la lista val_x. Hace el cálculo mediante la definición de las funciones por Series de Taylor y consecuente a esto, al valor que retorna
     la función le suma 1 y lo almacena en val_y. Al final retorna esta lista.
     Entradas
       funcion: Es una variable tipo string que contiene el nombre de una de las 4 funciones que recibe el programa: sen, cos, exp o ln.
       val_x: Es una lista de números flotantes que contiene valores de x para evaluar en la funcion
     Salidas
       val_y: Es una lista de números flotantes. Esta lista se llena evaluando los valores de x de la lista val_x que fueron enviados a alguna de las funciones que evaluan
       valores de x mediante las Series de Taylor: sen(x), cos(x), exp(x) o ln(x).
     Error
       * Si x entra en ln(x) y es menor a 0 (negativo), se mostrará un error
       * Si la lista val_x está vacía se mostrará un error
  '''
  val_y = []
  if funcion == "sen":
    val_y = [sen(x)+1 for x in val_x]
  elif funcion == "cos":
    val_y = [cos(x)+2 for x in val_x]
  elif funcion == "exp":
    val_y = [exp(x)-1 for x in val_x]
  elif funcion == "ln":
    val_y = [ln(x) for x in val_x]
  return val_y
```
